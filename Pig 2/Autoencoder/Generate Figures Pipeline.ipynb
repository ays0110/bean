```python
import sys
import numpy as np
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
import os
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import nibabel as nib
```

```python
#For doing figures by filters

parent_folder = 'Brandon Model'
gamma_string = 'Gamma 01'

mask = np.load('/home/chris/projects/rfmri/Data/pig_2/brandon/mask_2017_09_18.npy')
anatomical = nib.load('/home/chris/projects/rfmri/Data/pig_2/coregistered_anatomy.nii').get_data()
x,y,z = mask.nonzero()
slice_mask = ((z >= 7) & (z <= 16))
new_x = x[slice_mask]
new_y = y[slice_mask]
new_z = z[slice_mask] 

for cluster_i in range(2,6):
    for filter_i in range(32):
        labels = np.load("./" + parent_folder + '/' + gamma_string + '/filter_' + str(filter_i) + '_' + str(cluster_i) + '_clusters.npy')
        unique_labels = np.unique(labels)
        rgbs = [np.array(c) for c in sns.color_palette("bright", len(unique_labels))]
        assert(len(new_x) == len(labels))

        for label_i in range(len(unique_labels)):
            plt.figure(figsize=(40,20))
            label_of_interest = unique_labels[label_i]
            for slice_i in range(7,17):
                plt.subplot(2,5,slice_i-6)
                plt.imshow(anatomical[:,:,slice_i], cmap='viridis')
                for i in range(new_x.shape[0]):
                    if new_z[i] == slice_i and labels[i] == label_of_interest:
                        pass
                        plt.scatter(x=new_y[i], y=new_x[i], c='r', s=3)

            plt.savefig('./' + parent_folder + '/' + gamma_string + '/filter_' + str(filter_i) + '_'+ str(len(unique_labels)) + '_clusters_cluster_' + str(label_of_interest)+ '.png')
            plt.close('all')
```
